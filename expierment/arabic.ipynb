```python
import re
stop=nltk.corpus.stopwords.words("arabic")
def clean_text(text):
    text =re.sub(r'[^\w\s]', '', text) 
    text =re.sub("[إأآا]", "ا", text)
    text =re.sub("ى", "ي", text)
    words=text.split()
    words=[word for word in words if word not in stop]
    return " ".join(words)
   
df["text"]= df["text"].apply(clean_text)
```

```python
def commom_word_per_label(df):
    for l in df["label"].unique():
        print(f"Label: {l}")
        common_words = df[df["label"]==l]["text"].str.split(expand=True).stack().value_counts().head(5)
        print(common_words)
        print("-" * 50)
commom_word_per_label(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["label"]=le.fit_transform(df["label"])
```

```python
df.to_csv(r"D:\__Projects\LangFlow-Classifier\Data\Arabic\processed\ar_reviews_100k_cleaned.tsv",index=False,sep="\t")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(max_features=500)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
x=df["text"]
y=df["label"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train=vectorizer.fit_transform(x_train).toarray()
x_test=vectorizer.transform(x_test).toarray()
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
y_perd=gnb.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_perd))
```

```python
import joblib
joblib.dump(gnb,r"D:\__Projects\LangFlow-Classifier\models\arabic\arabic_sentiment_model.joblib")
joblib.dump(vectorizer,r"D:\__Projects\LangFlow-Classifier\models\arabic\arabic_vectorizer.joblib")
joblib.dump(le,r"D:\__Projects\LangFlow-Classifier\models\arabic\arabic_label_encoder.joblib")
```
