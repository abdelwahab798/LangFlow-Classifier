```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
arabic_df=pd.read_csv(r"D:\__Projects\LangFlow-Classifier\Data\Arabic\processed\ar_reviews_100k_cleaned.tsv",sep="\t")
english_df=pd.read_csv(r"D:\__Projects\LangFlow-Classifier\Data\English\processed\sentiment_data_cleaned.csv")
```

```python
arabic=arabic_df[["text"]].sample(10000, random_state=42).copy()
arabic["lang"]=0
english=english_df[["text"]].sample(10000, random_state=42).copy()
english["lang"]=1
df=pd.concat([arabic, english], ignore_index=True).sample(frac=1,random_state=42)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import joblib

x = df["text"]
y = df["lang"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(max_features=1000)
x_train_tfidf = vectorizer.fit_transform(x_train).toarray()
x_test_tfidf = vectorizer.transform(x_test).toarray()
gnb = GaussianNB()
gnb.fit(x_train_tfidf, y_train)
y_pred = gnb.predict(x_test_tfidf)
print(classification_report(y_test, y_pred))
```

```python
joblib.dump(gnb, r"D:\__Projects\LangFlow-Classifier\models\lang_classifier\lang_classifier_model.joblib")
joblib.dump(vectorizer, r"D:\__Projects\LangFlow-Classifier\models\lang_classifier\lang_vectorizer.joblib")
```
